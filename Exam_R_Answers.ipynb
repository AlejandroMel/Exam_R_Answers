# 8

```r
agg = aggregate(dfall[,2:4], by = list(dfall$CCAA), FUN = mean)
agg[3] <- NULL
agg
```

```r
agg1 = aggregate(dfall[,2:4], by = list(dfall$CCAA), FUN = sd)
agg1[3] <- NULL
agg1
```

# 9

```r
aggall = cbind(agg, agg1)
aggall[4] <- NULL
colnames(aggall) <- c('CCAA', 'GDP_mean', 'HDI_mean', 'GDP_sd', 'HDI_sd')
aggall
```
